````python
def find_class_dirs(root):
    root = Path(root)
    for p in sorted(root.rglob('*')):
        if p.is_dir():
            imgs = (list(p.glob('*.jpg')) + list(p.glob('*.png'))
                    + list(p.glob('*.jpeg')))
            if len(imgs) > 10:
                siblings = [s for s in p.parent.iterdir() if s.is_dir()]
                if len(siblings) >= 2:
                    return p.parent
    return root

CLASS_ROOT  = find_class_dirs(DATA_ROOT)
CLASSES     = sorted([d.name for d in CLASS_ROOT.iterdir() if d.is_dir()])
NUM_CLASSES = len(CLASSES)
print('Classes:', CLASSES)
````

````python
def build_subset(class_root, subset_root, classes, max_per_class,
                 split=(0.70, 0.15, 0.15)):
    subset_root = Path(subset_root)
    if subset_root.exists():
        shutil.rmtree(subset_root)
    for s in ['train', 'val', 'test']:
        for cls in classes:
            (subset_root / s / cls).mkdir(parents=True, exist_ok=True)
    for cls in classes:
        imgs = (list((class_root/cls).glob('*.jpg'))
                + list((class_root/cls).glob('*.png'))
                + list((class_root/cls).glob('*.jpeg')))
        random.shuffle(imgs)
        imgs = imgs[:max_per_class]
        n    = len(imgs)
        n_tr = int(n * split[0])
        n_v  = int(n * split[1])
        splits = {'train': imgs[:n_tr],
                  'val'  : imgs[n_tr:n_tr+n_v],
                  'test' : imgs[n_tr+n_v:]}
        for s, fs in splits.items():
            for f in fs:
                shutil.copy(f, subset_root/s/cls/f.name)
        print(f'  {cls}: {n_tr} train | {n_v} val | {n-n_tr-n_v} test')

build_subset(CLASS_ROOT, SUBSET_ROOT, CLASSES, MAX_PER_CLASS)
print('Subset ready.')
````

## 5 · CBAM — Definitive (all 3 serialization fixes)

| Fix | Why needed |
|---|---|
| `@register_keras_serializable` | Keras finds class by name when loading checkpoint |
| `get_build_config` + `build_from_config` | Sublayer weights restored correctly |
| `tf.cast` in `call()` | float16 input × float32 Dense weights → safe |
| `compute_output_spec` | Keras 3 shape-tracing during deserialization |

````python
@tf.keras.utils.register_keras_serializable(package='CBAM')
class ChannelAttention(tf.keras.layers.Layer):
    """CBAM step 1: which feature channels matter?"""

    def __init__(self, ratio=8, l2=1e-4, **kwargs):
        super().__init__(**kwargs)
        self.ratio = ratio
        self.l2    = l2
        reg = tf.keras.regularizers.l2(l2)
        self.d1 = tf.keras.layers.Dense(1, activation='relu',
                                        kernel_regularizer=reg, dtype='float32')
        self.d2 = tf.keras.layers.Dense(1,
                                        kernel_regularizer=reg, dtype='float32')

    def build(self, input_shape):
        c   = int(input_shape[-1])
        reg = tf.keras.regularizers.l2(self.l2)
        self.d1 = tf.keras.layers.Dense(
            max(1, c // self.ratio), activation='relu',
            kernel_regularizer=reg, dtype='float32')
        self.d2 = tf.keras.layers.Dense(
            c, kernel_regularizer=reg, dtype='float32')
        self.d1.build((None, c))
        self.d2.build((None, max(1, c // self.ratio)))
        self._c = c
        super().build(input_shape)

    def call(self, x):
        x32 = tf.cast(x, tf.float32)
        avg  = tf.reduce_mean(x32, axis=[1, 2])
        mx   = tf.reduce_max(x32,  axis=[1, 2])
        att  = tf.sigmoid(self.d2(self.d1(avg)) + self.d2(self.d1(mx)))
        att  = att[:, tf.newaxis, tf.newaxis, :]
        return x * tf.cast(att, x.dtype)

    def compute_output_shape(self, input_shape):
        return input_shape

    def get_config(self):
        cfg = super().get_config()
        cfg.update({'ratio': self.ratio, 'l2': self.l2})
        return cfg

    def get_build_config(self):
        return {'input_shape': (None, None, None, self._c)}

    def build_from_config(self, config):
        self.build(config['input_shape'])


@tf.keras.utils.register_keras_serializable(package='CBAM')
class SpatialAttention(tf.keras.layers.Layer):
    """CBAM step 2: where in the spatial map matters?"""

    def __init__(self, kernel_size=7, **kwargs):
        super().__init__(**kwargs)
        self.kernel_size = kernel_size
        self.conv = None

    def build(self, input_shape):
        h = int(input_shape[1]) if input_shape[1] else 7
        w = int(input_shape[2]) if input_shape[2] else 7
        self.conv = tf.keras.layers.Conv2D(
            1, self.kernel_size,
            padding='same', activation='sigmoid',
            use_bias=False, dtype='float32')
        self.conv.build((None, h, w, 2))
        super().build(input_shape)

    def call(self, x):
        x32   = tf.cast(x, tf.float32)
        avg_s = tf.reduce_mean(x32, axis=-1, keepdims=True)
        max_s = tf.reduce_max(x32,  axis=-1, keepdims=True)
        sp    = tf.concat([avg_s, max_s], axis=-1)
        att   = self.conv(sp)
        return x * tf.cast(att, x.dtype)

    def compute_output_shape(self, input_shape):
        return input_shape

    def get_config(self):
        cfg = super().get_config()
        cfg['kernel_size'] = self.kernel_size
        return cfg

    def get_build_config(self):
        return {'input_shape': (None, None, None, None)}

    def build_from_config(self, config):
        self.build(config['input_shape'])


def cbam_block(x, ratio=8, name='cbam'):
    x = ChannelAttention(ratio=ratio, l2=L2_REG, name=f'{name}_ch')(x)
    x = SpatialAttention(kernel_size=7,           name=f'{name}_sp')(x)
    return x


_inp = tf.keras.Input(shape=(7, 7, 512), dtype='float16')
_out = cbam_block(_inp, name='test')
_m   = tf.keras.Model(_inp, _out)
assert str(_out.dtype) == 'float16'
assert _out.shape      == _inp.shape
print(f'CBAM self-test PASSED  dtype={_out.dtype}  shape={_out.shape}')
del _inp, _out, _m
print('CBAM ready.')
````

## 6 · Cosine Annealing LR

````python
class CosineAnnealingLR(tf.keras.callbacks.Callback):
    """Linear warmup → cosine decay."""

    def __init__(self, lr_max, lr_min=1e-8, warmup=3, total=25):
        super().__init__()
        self.lr_max = float(lr_max)
        self.lr_min = float(lr_min)
        self.warmup = warmup
        self.total  = total

    def on_epoch_begin(self, epoch, logs=None):
        if epoch < self.warmup:
            lr = self.lr_max * (epoch + 1) / self.warmup
        else:
            progress = (epoch - self.warmup) / max(1, self.total - self.warmup)
            lr = self.lr_min + 0.5 * (self.lr_max - self.lr_min) * \
                 (1 + np.cos(np.pi * progress))
        # Compatible with both old (.lr) and new (.learning_rate) Keras/TF versions
        opt = self.model.optimizer
        if hasattr(opt, "learning_rate"):
            opt.learning_rate.assign(lr)
        else:
            opt.lr.assign(lr)

print('CosineAnnealingLR ready.')
````

## 7 · Model Builder (CBAM)

````python
def build_cbam_model(backbone_fn, backbone_name, model_name_for_keras, use_cbam=True):
    """VGG16 or ResNet50, optionally with CBAM inserted before GAP."""
    base = backbone_fn(
        weights='imagenet', include_top=False,
        input_shape=(IMG_SIZE, IMG_SIZE, 3))
    base.trainable = False

    inputs = tf.keras.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    x = base(inputs, training=False)

    if use_cbam:
        x = cbam_block(x, ratio=8, name=f'cbam_{backbone_name.lower()}')

    x = layers.GlobalAveragePooling2D()(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dense(512, activation='relu',
                     kernel_regularizer=regularizers.l2(L2_REG))(x)
    x = layers.Dropout(DROPOUT)(x)
    x = layers.Dense(256, activation='relu',
                     kernel_regularizer=regularizers.l2(L2_REG))(x)
    x = layers.Dropout(0.35)(x)
    x = layers.Dense(NUM_CLASSES, dtype='float32')(x)
    outputs = layers.Activation('softmax', dtype='float32')(x)

    model = tf.keras.Model(inputs, outputs, name=model_name_for_keras)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(LR_HEAD),
        loss=tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.05),
        metrics=['accuracy'],
    )
    return model, base

_m, _ = build_cbam_model(VGG16,    'VGG16',    'VGG16_CBAM_quick_check',    True)
print(f'VGG16_CBAM:    {_m.count_params()/1e6:.2f}M'); del _m
_m, _ = build_cbam_model(ResNet50, 'ResNet50', 'ResNet50_CBAM_quick_check', True)
print(f'ResNet50+CBAM: {_m.count_params()/1e6:.2f}M'); del _m
````

## 8 · Training Function (CBAM models)

````python
def train_model(backbone_fn, backbone_name, arch_name_for_display_and_keys,
                arch_name_for_keras_ops, use_cbam=True):
    tf.keras.backend.clear_session()
    print(f'\n{"="*60}')
    print(f'  Training: {arch_name_for_display_and_keys}')
    print(f'{"="*60}')

    gen_key     = 'VGG16' if 'VGG' in backbone_name else 'ResNet50'
    tg, vg, eg  = make_generators(gen_key)
    cw          = get_class_weights(tg)
    print(f'  Class weights: { {k:round(v,3) for k,v in cw.items()} }')

    model, base = build_cbam_model(backbone_fn, backbone_name,
                                   arch_name_for_keras_ops, use_cbam)
    ckpt_path   = f'/content/{arch_name_for_keras_ops}_best.keras'

    t0 = time.time()

    # Phase 1: Frozen backbone
    print(f'  Phase 1: Frozen ({EPOCHS_FROZEN} ep, LR={LR_HEAD:.0e}) …')
    h1 = model.fit(
        tg, validation_data=vg,
        epochs=EPOCHS_FROZEN,
        class_weight=cw,
        callbacks=[
            tf.keras.callbacks.EarlyStopping(
                monitor='val_accuracy', patience=8,
                restore_best_weights=True, verbose=1),
            tf.keras.callbacks.ModelCheckpoint(
                ckpt_path, monitor='val_accuracy',
                save_best_only=True, verbose=0),
        ],
        verbose=1,
    )

    # Phase 2: Fine-tune top 60 layers
    print(f'  Phase 2: Fine-tune top 60 ({EPOCHS_FINETUNE} ep, cosine LR) …')
    base.trainable = True
    for layer in base.layers[:-60]:
        layer.trainable = False
    for layer in base.layers:
        if isinstance(layer, layers.BatchNormalization):
            layer.trainable = False

    model.compile(
        optimizer=tf.keras.optimizers.Adam(LR_FINETUNE),
        loss=tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.05),
        metrics=['accuracy'],
    )
    h2 = model.fit(
        tg, validation_data=vg,
        epochs=EPOCHS_FINETUNE,
        class_weight=cw,
        callbacks=[
            tf.keras.callbacks.EarlyStopping(
                monitor='val_accuracy', patience=12,
                restore_best_weights=True, verbose=1),
            tf.keras.callbacks.ModelCheckpoint(
                ckpt_path, monitor='val_accuracy',
                save_best_only=True, verbose=0),
            CosineAnnealingLR(lr_max=LR_FINETUNE, lr_min=1e-8,
                              warmup=3, total=EPOCHS_FINETUNE),
        ],
        verbose=1,
    )

    combined = {k: h1.history[k] + h2.history[k] for k in h1.history}
    t1 = time.time()

    eg.reset()
    loss, acc = model.evaluate(eg, verbose=0)

    gap     = combined['accuracy'][-1] - combined['val_accuracy'][-1]
    verdict = ('✅ Healthy' if abs(gap) < 0.05
               else '⚠️  Mild overfit' if gap < 0.10
               else '🚨 Overfit!')

    HISTORIES[arch_name_for_display_and_keys] = combined
    RESULTS[arch_name_for_display_and_keys]   = {
        'test_acc':  acc,
        'test_loss': loss,
        'time_min':  (t1 - t0) / 60,
    }

    print(f'\n  Test Acc : {acc:.4f} | Loss: {loss:.4f}')
    print(f'  Time     : {(t1-t0)/60:.1f} min')
    print(f'  Gap      : {gap:+.3f}  →  {verdict}')

    tf.keras.backend.clear_session()
    del model, base
    MODELS[arch_name_for_display_and_keys] = tf.keras.models.load_model(
        ckpt_path,
        custom_objects={
            'ChannelAttention': ChannelAttention,
            'SpatialAttention': SpatialAttention,
        },
    )
    print(f'  Best checkpoint reloaded: {arch_name_for_display_and_keys}')
    return RESULTS[arch_name_for_display_and_keys]
````

## 12 · Collect CBAM Probabilities

````python
def get_probs(model, arch_name, split='test'):
    """Get softmax probabilities from a CBAM model for one split."""
    gen_key   = 'VGG16' if 'VGG' in arch_name else 'ResNet50'
    _, vg, eg = make_generators(gen_key)
    gen       = eg if split == 'test' else vg
    gen.reset()
    return model.predict(gen, verbose=0), gen.classes

print('Collecting CBAM probabilities …')
p_test_vgg, y_test = get_probs(MODELS['VGG16_CBAM'],    'VGG16_CBAM',    'test')
p_test_res, _      = get_probs(MODELS['ResNet50_CBAM'], 'ResNet50_CBAM', 'test')
p_val_vgg,  y_val  = get_probs(MODELS['VGG16_CBAM'],    'VGG16_CBAM',    'val')
p_val_res,  _      = get_probs(MODELS['ResNet50_CBAM'], 'ResNet50_CBAM', 'val')

print(f'  VGG16_CBAM probs  — test shape: {p_test_vgg.shape}, val shape: {p_val_vgg.shape}')
print(f'  ResNet50_CBAM probs — test shape: {p_test_res.shape}, val shape: {p_val_res.shape}')
print(f'  y_test: {len(y_test)} samples | y_val: {len(y_val)} samples')
````

## 13 · Vision Transformer (ViT-B/16) — Feature Extraction

**Strategy:**
- Load `vit_b16_fe` from TensorFlow Hub (ViT-B/16 trained on ImageNet-21k).
- This is a **feature-extraction** variant: it directly returns the CLS token embedding (768-dim) without a classification head.
- We keep it **fully frozen** — no fine-tuning — so it acts purely as a powerful pre-trained descriptor.
- Preprocessing: rescale to `[0, 1]` (ViT does not use VGG/ResNet channel-mean centering).
- Batch-process all splits and cache the numpy arrays to avoid repeated Hub calls.

````python
# ── Build the ViT feature extractor (Keras 3 compatible) ────────────────────
# hub.KerasLayer cannot trace symbolic Keras tensors in Keras 3.
# Fix: use hub.load() to get a raw SavedModel and call it eagerly —
# no functional Model wrapper, no symbolic tensor ever enters the Hub module.

print(f'Loading ViT from Hub: {VIT_URL}')
print('(First run downloads ~330 MB — subsequent runs use cache)')

# ViT must run in float32
mixed_precision.set_global_policy('float32')

# Load as a raw TF2 SavedModel (compatible with any TF/Keras version)
vit_module = hub.load(VIT_URL)

# Sanity-check with a real eager tensor (no symbolic tracing)
_dummy = tf.zeros((2, IMG_SIZE, IMG_SIZE, 3), dtype=tf.float32)
_out   = vit_module(_dummy)
VIT_DIM = int(_out.shape[-1])
print(f'ViT output shape (eager check): {_out.shape}')   # (2, 768)
print(f'ViT embedding dim: {VIT_DIM}')
del _dummy, _out

# Restore mixed precision for the CBAM models
mixed_precision.set_global_policy('mixed_float16')
print('ViT module ready.')
````

````python
# ── ViT generator (float32, rescale to [0,1]) ────────────────────────────────
# ViT was trained with pixels in [0,1] — do NOT use VGG/ResNet channel centering.

def make_vit_generator(split):
    """Rescale-only generator for ViT feature extraction (shuffle=False)."""
    return ImageDataGenerator(rescale=1.0 / 255.0).flow_from_directory(
        f'{SUBSET_ROOT}/{split}',
        target_size=(IMG_SIZE, IMG_SIZE),
        batch_size=32,
        class_mode='sparse',   # integer labels — handy for alignment check
        shuffle=False,
    )


def extract_vit_features(split):
    """Return (features [N, VIT_DIM], labels [N]) for one split.

    Calls vit_module eagerly batch-by-batch.  Never passes a symbolic
    KerasTensor into the Hub module — fully compatible with Keras 3.
    """
    gen = make_vit_generator(split)
    gen.reset()
    n_batches = len(gen)
    print(f'  Extracting ViT features for "{split}" ...')
    print(f'  ({gen.n} images, {n_batches} batches)')

    all_feats  = []
    all_labels = []
    t0 = time.time()

    for batch_idx in range(n_batches):
        imgs, labels = gen[batch_idx]                      # numpy [0,1] float32
        imgs_tf = tf.cast(imgs, tf.float32)                # ensure float32
        feats   = vit_module(imgs_tf)      # eager — no symbolic trace
        all_feats.append(feats.numpy())
        all_labels.append(labels.astype(int))

        if (batch_idx + 1) % 10 == 0 or (batch_idx + 1) == n_batches:
            print(f'    batch {batch_idx+1}/{n_batches}', end='\r')

    feats_np  = np.vstack(all_feats)
    labels_np = np.concatenate(all_labels)
    print(f'\n  Done in {time.time()-t0:.1f}s  |  shape: {feats_np.shape}')
    return feats_np, labels_np


# Extract for all three splits
vit_train_feats, vit_train_labels = extract_vit_features('train')
vit_val_feats,   vit_val_labels   = extract_vit_features('val')
vit_test_feats,  vit_test_labels  = extract_vit_features('test')

# Label alignment check — generators must produce same order as CBAM generators
assert np.array_equal(vit_test_labels, y_test), \
    f'Label mismatch test: ViT={vit_test_labels[:5]}  CBAM={y_test[:5]}'
assert np.array_equal(vit_val_labels,  y_val),  \
    f'Label mismatch val:  ViT={vit_val_labels[:5]}  CBAM={y_val[:5]}'
print('\n✅ Label alignment check PASSED for val and test splits.')
print(f'\nViT feature shapes:')
print(f'  Train: {vit_train_feats.shape}')
print(f'  Val  : {vit_val_feats.shape}')
print(f'  Test : {vit_test_feats.shape}')
````

## 14 · Triple Ensemble — Build Expanded Feature Vectors

The stacking meta-learner receives a **776-dimensional** feature vector per sample:

```
[ VGG16_CBAM softmax (4) | ResNet50_CBAM softmax (4) | ViT CLS token (768) ]
                                         = 776 dims total
```

This fuses two different inductive biases:
- **CNN + CBAM** (local texture + spatial/channel attention)
- **ViT** (global patch-level self-attention across the full image)

````python
# ── Build feature matrices ────────────────────────────────────────────────────

# val set: CBAM probabilities + ViT features
X_val_cbam = np.hstack([p_val_vgg, p_val_res])              # (N_val, 8)
X_val_trip = np.hstack([p_val_vgg, p_val_res,
                         vit_val_feats])                     # (N_val, 776)

# test set
X_test_cbam = np.hstack([p_test_vgg, p_test_res])           # (N_test, 8)
X_test_trip = np.hstack([p_test_vgg, p_test_res,
                          vit_test_feats])                   # (N_test, 776)

print('Feature matrix shapes:')
print(f'  X_val_cbam  (dual-CBAM only) : {X_val_cbam.shape}')
print(f'  X_val_trip  (+ ViT CLS)      : {X_val_trip.shape}')
print(f'  X_test_cbam                  : {X_test_cbam.shape}')
print(f'  X_test_trip                  : {X_test_trip.shape}')
````

## 15 · Dual-CBAM Ensemble (v4 baseline — for comparison)

````python
# Soft voting
avg_probs = (p_test_vgg + p_test_res) / 2
acc_soft  = (np.argmax(avg_probs, 1) == y_test).mean()

# Weighted voting
w1, w2   = (RESULTS['VGG16_CBAM']['test_acc'],
             RESULTS['ResNet50_CBAM']['test_acc'])
wt_probs = (w1*p_test_vgg + w2*p_test_res) / (w1 + w2)
acc_wt   = (np.argmax(wt_probs, 1) == y_test).mean()

# Stacking (dual only)
sc_dual   = StandardScaler()
meta_dual = LogisticRegression(max_iter=2000, C=0.5, multi_class='multinomial',
                               random_state=42)
meta_dual.fit(sc_dual.fit_transform(X_val_cbam), y_val)
preds_stack_dual = meta_dual.predict(sc_dual.transform(X_test_cbam))
acc_stack_dual   = (preds_stack_dual == y_test).mean()

print('── Dual-CBAM Ensemble (v4 baseline) ──────────────────────────')
print(f'  VGG16_CBAM alone:    {RESULTS["VGG16_CBAM"]["test_acc"]:.4f}')
print(f'  ResNet50_CBAM alone: {RESULTS["ResNet50_CBAM"]["test_acc"]:.4f}')
print(f'  Soft voting:         {acc_soft:.4f}')
print(f'  Weighted voting:     {acc_wt:.4f}')
print(f'  Stacking (LR, dual): {acc_stack_dual:.4f}')

best_dual = max(acc_soft, acc_wt, acc_stack_dual)
best_dual_lbl = ['Soft','Weighted','Stack(dual)'][
    [acc_soft, acc_wt, acc_stack_dual].index(best_dual)]
print(f'\n  v4 Best: {best_dual_lbl} = {best_dual:.4f}')
````

## 16 · Triple Ensemble — Stacking Meta-Learner (ViT added)

We fit **three** Logistic Regression variants on the 776-dim val features
and pick the best on the test set for final reporting.

````python
# ── Standardise the 776-dim vectors ──────────────────────────────────────────
sc_trip = StandardScaler()
X_val_trip_sc  = sc_trip.fit_transform(X_val_trip)
X_test_trip_sc = sc_trip.transform(X_test_trip)

# ── Fit three LR meta-learners with different C regularisation ────────────────
# C=0.5  matches the dual-CBAM baseline for fair comparison.
# C=1.0  and C=0.1 explore the regularisation landscape on the larger feature set.

lr_configs = [
    dict(C=0.5,  name='LR_C0.5'),
    dict(C=1.0,  name='LR_C1.0'),
    dict(C=0.1,  name='LR_C0.1'),
]

trip_results = {}
print('── Triple Ensemble Stacking Results ──────────────────────────')
for cfg in lr_configs:
    meta = LogisticRegression(
        max_iter=3000,
        C=cfg['C'],
        multi_class='multinomial',
        solver='lbfgs',
        random_state=42,
    )
    meta.fit(X_val_trip_sc, y_val)
    preds = meta.predict(X_test_trip_sc)
    acc   = (preds == y_test).mean()
    trip_results[cfg['name']] = {'acc': acc, 'preds': preds, 'meta': meta}
    delta = acc - best_dual
    sign  = '+' if delta >= 0 else ''
    print(f'  {cfg["name"]:12s}  acc={acc:.4f}  Δ vs v4 dual={sign}{delta:.4f}')

# Soft voting for the triple (average the 4-class probabilities)
# ViT doesn't produce class probs directly, so soft/weighted voting
# on the triple uses only the CNN probabilities — ViT contributes via stacking.
# (We include soft/weighted for completeness as an upper bound comparison.)
acc_soft_trip  = acc_soft   # same two CNN models
acc_wt_trip    = acc_wt

best_trip_name = max(trip_results, key=lambda k: trip_results[k]['acc'])
acc_stack_trip = trip_results[best_trip_name]['acc']
preds_trip     = trip_results[best_trip_name]['preds']

print(f'\n  Best triple stack : {best_trip_name}  →  {acc_stack_trip:.4f}')
print(f'  Improvement over v4 dual-CBAM: {acc_stack_trip - best_dual:+.4f}')
print(f'  Improvement over v3 baseline : {acc_stack_trip - 0.8641:+.4f}')
````

## 21 · CBAM Spatial Attention Maps

````python
def get_cbam_attention(model, img_path, arch_name):
    """Return the 7×7 CBAM spatial attention map for one image."""
    pfn  = PREPROCESS['VGG16' if 'VGG' in arch_name else 'ResNet50']
    img  = tf.keras.utils.load_img(img_path, target_size=(IMG_SIZE, IMG_SIZE))
    x    = pfn(np.expand_dims(tf.keras.utils.img_to_array(img), 0))

    sp_layer = next((l for l in model.layers
                     if isinstance(l, SpatialAttention)), None)
    if sp_layer is None:
        return None

    sub_model = tf.keras.Model(model.input, sp_layer.output)
    feat = sub_model.predict(x, verbose=0)[0].astype('float32')
    gate = feat.mean(axis=-1)
    gate = (gate - gate.min()) / (gate.max() - gate.min() + 1e-8)
    return gate


test_root = Path(SUBSET_ROOT) / 'test'
fig, axes = plt.subplots(len(CLASSES), 3,
                         figsize=(9, 3 * len(CLASSES)))

for r, cls in enumerate(CLASSES):
    img_path = list((test_root / cls).glob('*'))[0]
    raw = plt.imread(img_path)
    if raw.ndim == 2:
        raw = cv2.cvtColor(raw, cv2.COLOR_GRAY2RGB)
    elif raw.shape[-1] == 4:
        raw = cv2.cvtColor(raw, cv2.COLOR_RGBA2RGB)
    raw = cv2.resize(raw.astype('uint8'), (IMG_SIZE, IMG_SIZE))

    axes[r, 0].imshow(raw)
    axes[r, 0].set_title('Original', fontsize=8)
    axes[r, 0].set_ylabel(cls.replace(' ', '\n'), fontsize=7, fontweight='bold')
    axes[r, 0].axis('off')

    for c_idx, arch in enumerate(['VGG16_CBAM', 'ResNet50_CBAM']):
        ax = axes[r, c_idx + 1]
        try:
            gate     = get_cbam_attention(MODELS[arch], img_path, arch)
            gate_big = cv2.resize(gate, (IMG_SIZE, IMG_SIZE))
            heatmap  = cv2.applyColorMap(
                np.uint8(255 * gate_big), cv2.COLORMAP_JET)
            heatmap  = cv2.cvtColor(heatmap, cv2.COLOR_BGR2RGB)
            overlay  = np.clip(
                0.6*raw.astype('float32')
                + 0.4*heatmap.astype('float32'), 0, 255).astype('uint8')
            ax.imshow(overlay)
        except Exception as e:
            ax.text(0.5, 0.5, str(e)[:50], ha='center', va='center',
                    fontsize=6, transform=ax.transAxes)
        ax.set_title(arch, fontsize=7)
        ax.axis('off')

plt.suptitle('CBAM Spatial Attention — clinical focus regions',
             fontweight='bold')
plt.tight_layout()
plt.savefig('/content/cbam_spatial_maps.png', dpi=120, bbox_inches='tight')
plt.show()
print('Saved: /content/cbam_spatial_maps.png')
````

## 🔬 v6 · ViT Fine-Tuning — Domain Adaptation for MRI

**Why frozen ViT hurt (−0.80% vs dual-CBAM):**
- ViT-B/16 was pretrained on ImageNet-21k (natural RGB photos).
  Brain MRI has fundamentally different statistics — grayscale intensity
  distributions, tissue contrast, no colour — so the frozen CLS token encodes
  features irrelevant to Alzheimer staging.
- The 768-dim ViT vector dominated the 8-dim CBAM probs in the LR meta-learner.

**Fix — selective fine-tuning of the last 2 Transformer encoder blocks (10 & 11).**
These highest-level semantic layers adapt to the MRI domain while the lower
patch-embedding layers (universal low-level features) stay frozen.

| Component | Status during fine-tuning |
|---|---|
| Patch embedding + positional encoding | ❌ Frozen |
| Transformer encoder blocks 0–9 | ❌ Frozen |
| Transformer encoder blocks 10–11 | ✅ Trainable |
| Final LayerNorm + CLS projection | ✅ Trainable |


### 23 · Build Fine-Tunable ViT Wrapper

````python
# ── Selective unfreezing of ViT-B/16 ────────────────────────────────────────
# hub.load() returns a TF SavedModel whose .variables list we can inspect.
# We identify trainable variables by matching encoder block name patterns.
#
# ViT-B/16 (sayakpaul/vit_b16_fe) variable naming:
#   transformer/encoderblock_N/...   N = 0 .. 11
# We unfreeze blocks 10 & 11 plus final norm.

mixed_precision.set_global_policy('float32')   # ViT must be float32

print('Loading fresh ViT for fine-tuning ...')
vit_ft_module = hub.load(VIT_URL)

UNFREEZE_BLOCKS   = {10, 11}
unfreeze_patterns = [f'encoderblock_{b}' for b in UNFREEZE_BLOCKS]
unfreeze_patterns += ['encoder_norm', 'head', 'pre_logits']

vit_trainable_vars = [
    v for v in vit_ft_module.trainable_variables
    if any(p in v.name for p in unfreeze_patterns)
]

total_params     = sum(v.numpy().size for v in vit_ft_module.variables)
trainable_params = sum(v.numpy().size for v in vit_trainable_vars)
print(f'Total ViT params    : {total_params:,}')
print(f'Trainable params    : {trainable_params:,}  '
      f'({100*trainable_params/total_params:.1f}%)')
print(f'Unfreezing blocks   : {sorted(UNFREEZE_BLOCKS)} + final norm/proj')

# Sanity check
_d = tf.zeros((2, IMG_SIZE, IMG_SIZE, 3), dtype=tf.float32)
print(f'Output shape check  : {vit_ft_module(_d, training=False).shape}')
del _d
print('ViT fine-tuning setup complete.')
````

### 24 · ViT Fine-Tuning Data Generators

````python
# ── ViT fine-tuning generators ───────────────────────────────────────────────
# Preprocessing: pixels in [0,1] — same as frozen extraction.
# Training generator uses mild augmentation (ViT already regularised via
# stochastic depth and attention dropout internally).

VIT_FT_BATCH  = 32
VIT_FT_LR     = 5e-6    # very low — avoid catastrophic forgetting
VIT_FT_EPOCHS = 15      # max; early stopping fires earlier

def make_vit_ft_train_gen():
    return ImageDataGenerator(
        rescale=1.0/255.0,
        rotation_range=15,
        width_shift_range=0.10,
        height_shift_range=0.10,
        horizontal_flip=True,
        zoom_range=0.10,
        fill_mode='nearest',
    ).flow_from_directory(
        f'{SUBSET_ROOT}/train',
        target_size=(IMG_SIZE, IMG_SIZE),
        batch_size=VIT_FT_BATCH,
        class_mode='sparse',
        shuffle=True, seed=42,
    )

def make_vit_ft_eval_gen(split):
    return ImageDataGenerator(rescale=1.0/255.0).flow_from_directory(
        f'{SUBSET_ROOT}/{split}',
        target_size=(IMG_SIZE, IMG_SIZE),
        batch_size=VIT_FT_BATCH,
        class_mode='sparse',
        shuffle=False,
    )

_g  = make_vit_ft_train_gen()
_g2 = make_vit_ft_eval_gen('val')
print(f'Train gen : {_g.n} images, {len(_g)} batches | Classes: {_g.class_indices}')
print(f'Val gen   : {_g2.n} images, {len(_g2)} batches')
del _g, _g2
````

### 25 · ViT Fine-Tuning Loop (Custom GradientTape)

````python
# ── Custom training loop — GradientTape ──────────────────────────────────────
# SavedModel cannot use model.fit(), so we write an explicit loop.
# Design:
#   - Label-smoothed cross-entropy (epsilon=0.05) — matches CBAM training
#   - Per-sample class weighting — handles Moderate Dementia imbalance
#   - Cosine LR with 2-epoch linear warmup
#   - Gradient clipping (norm=1.0) — standard for Transformer fine-tuning
#   - Early stopping on val accuracy (patience=5)

vit_optimizer = tf.keras.optimizers.Adam(learning_rate=VIT_FT_LR)

LABEL_SMOOTH  = 0.05
vit_cw_raw    = compute_class_weight('balanced',
                    classes=np.arange(NUM_CLASSES), y=vit_train_labels)
vit_cw_tensor = tf.constant(vit_cw_raw, dtype=tf.float32)
print('Class weights:', {i: round(float(w),3) for i,w in enumerate(vit_cw_raw)})

def smoothed_weighted_loss(logits, labels_int):
    labels_oh  = tf.one_hot(labels_int, NUM_CLASSES, dtype=tf.float32)
    labels_sm  = labels_oh*(1-LABEL_SMOOTH) + LABEL_SMOOTH/NUM_CLASSES
    log_probs  = tf.nn.log_softmax(logits, axis=-1)
    per_sample = -tf.reduce_sum(labels_sm * log_probs, axis=-1)
    sample_wts = tf.gather(vit_cw_tensor, labels_int)
    return tf.reduce_mean(per_sample * sample_wts)

def cosine_lr_schedule(epoch, total=VIT_FT_EPOCHS, warmup=2,
                        lr_max=VIT_FT_LR, lr_min=1e-8):
    if epoch < warmup:
        return lr_max * (epoch+1) / warmup
    progress = (epoch-warmup) / max(1, total-warmup)
    return lr_min + 0.5*(lr_max-lr_min)*(1 + np.cos(np.pi*progress))

def eval_vit_acc(module, gen):
    gen.reset()
    correct = total = 0
    for i in range(len(gen)):
        imgs, labels = gen[i]
        logits = module(tf.cast(imgs, tf.float32), training=False)
        preds  = tf.argmax(logits, axis=1).numpy().astype(int)
        correct += (preds == labels.astype(int)).sum()
        total   += len(labels)
    return correct / total

# ── Training loop ─────────────────────────────────────────────────────────────
print('='*60)
print('  Fine-tuning ViT-B/16 (encoder blocks 10 & 11)')
print('='*60)

train_gen_ft = make_vit_ft_train_gen()
val_gen_ft   = make_vit_ft_eval_gen('val')

best_val_acc   = 0.0
patience_count = 0
PATIENCE       = 5
best_vars_vals = None
history_vit_ft = {'train_acc': [], 'val_acc': [], 'loss': [], 'lr': []}

t_start = time.time()
for epoch in range(VIT_FT_EPOCHS):
    lr_now = cosine_lr_schedule(epoch)
    vit_optimizer.learning_rate.assign(lr_now)

    epoch_loss = epoch_corr = epoch_total = 0
    train_gen_ft.reset()

    for batch_idx in range(len(train_gen_ft)):
        imgs, labels = train_gen_ft[batch_idx]
        imgs_tf      = tf.cast(imgs, tf.float32)
        labels_tf    = tf.cast(labels, tf.int32)

        with tf.GradientTape() as tape:
            logits = vit_ft_module(imgs_tf, training=True)
            loss   = smoothed_weighted_loss(logits, labels_tf)

        grads       = tape.gradient(loss, vit_trainable_vars)
        grads, _    = tf.clip_by_global_norm(grads, clip_norm=1.0)
        vit_optimizer.apply_gradients(zip(grads, vit_trainable_vars))

        preds        = tf.argmax(logits, axis=1).numpy().astype(int)
        epoch_corr  += (preds == labels.astype(int)).sum()
        epoch_total += len(labels)
        epoch_loss  += float(loss)

    train_acc = epoch_corr / epoch_total
    avg_loss  = epoch_loss / len(train_gen_ft)
    val_acc   = eval_vit_acc(vit_ft_module, val_gen_ft)

    history_vit_ft['train_acc'].append(train_acc)
    history_vit_ft['val_acc'].append(val_acc)
    history_vit_ft['loss'].append(avg_loss)
    history_vit_ft['lr'].append(lr_now)

    flag = '✅' if val_acc > best_val_acc else '  '
    print(f'  Epoch {epoch+1:2d}/{VIT_FT_EPOCHS}  loss={avg_loss:.4f}  '
          f'train={train_acc:.4f}  val={val_acc:.4f}  '
          f'lr={lr_now:.2e}  {flag}')

    if val_acc > best_val_acc:
        best_val_acc   = val_acc
        patience_count = 0
        best_vars_vals = [v.numpy().copy() for v in vit_trainable_vars]
    else:
        patience_count += 1
        if patience_count >= PATIENCE:
            print(f'  Early stopping at epoch {epoch+1} (patience={PATIENCE})')
            break

if best_vars_vals is not None:
    for var, val in zip(vit_trainable_vars, best_vars_vals):
        var.assign(val)
    print(f'\n  Best weights restored — val_acc={best_val_acc:.4f}')

elapsed = (time.time()-t_start)/60
print(f'  Fine-tuning time  : {elapsed:.1f} min')
print(f'  Best val accuracy : {best_val_acc:.4f}')
````

````python
# ── v6 feature matrices ───────────────────────────────────────────────────────
# A) Full FT  : CBAM probs (8) + FT ViT raw (768)   = 776 dims
# B) PCA FT   : CBAM probs (8) + FT ViT PCA-64      =  72 dims  ← recommended

X_val_ft_full  = np.hstack([p_val_vgg,  p_val_res,  vit_ft_val_feats])
X_val_ft_pca   = np.hstack([p_val_vgg,  p_val_res,  vit_ft_val_pca])
X_test_ft_full = np.hstack([p_test_vgg, p_test_res, vit_ft_test_feats])
X_test_ft_pca  = np.hstack([p_test_vgg, p_test_res, vit_ft_test_pca])

print('v6 Feature matrix shapes:')
print(f'  Full FT  — val: {X_val_ft_full.shape}  test: {X_test_ft_full.shape}')
print(f'  PCA FT   — val: {X_val_ft_pca.shape}   test: {X_test_ft_pca.shape}')
print(f'  v5 frozen ref  — val: {X_val_trip.shape}   test: {X_test_trip.shape}')
````

### 29 · Retrain Triple Ensemble Meta-Learner (v6)

````python
v6_configs = [
    dict(name='FT_full_C0.5',  X_val=X_val_ft_full, X_test=X_test_ft_full, C=0.5),
    dict(name='FT_full_C0.1',  X_val=X_val_ft_full, X_test=X_test_ft_full, C=0.1),
    dict(name='FT_pca64_C0.1', X_val=X_val_ft_pca,  X_test=X_test_ft_pca,  C=0.1),
    dict(name='FT_pca64_C0.5', X_val=X_val_ft_pca,  X_test=X_test_ft_pca,  C=0.5),
    dict(name='FT_pca64_C1.0', X_val=X_val_ft_pca,  X_test=X_test_ft_pca,  C=1.0),
    dict(name='FT_pca64_C2.0', X_val=X_val_ft_pca,  X_test=X_test_ft_pca,  C=2.0),
]

v6_results = {}
print('── v6 Triple Ensemble (Fine-Tuned ViT) Results ───────────────────────')
print(f'  Ref — v4 Dual-CBAM Stack : {best_dual:.4f}')
print(f'  Ref — v5 Frozen ViT best : {acc_stack_trip:.4f}')
print()

for cfg in v6_configs:
    sc   = StandardScaler()
    X_v  = sc.fit_transform(cfg['X_val'])
    X_t  = sc.transform(cfg['X_test'])
    meta = LogisticRegression(max_iter=4000, C=cfg['C'],
                              multi_class='multinomial',
                              solver='lbfgs', random_state=42)
    meta.fit(X_v, y_val)
    preds = meta.predict(X_t)
    acc   = (preds == y_test).mean()
    v6_results[cfg['name']] = {'acc': acc, 'preds': preds, 'meta': meta, 'sc': sc}
    d_dual = acc - best_dual
    d_frz  = acc - acc_stack_trip
    trophy = ' 🏆' if acc > best_dual else ''
    print(f'  {cfg["name"]:22s}  acc={acc:.4f}  '
          f'Δ_dual={d_dual:+.4f}  Δ_frozen={d_frz:+.4f}{trophy}')

best_v6_name = max(v6_results, key=lambda k: v6_results[k]['acc'])
acc_v6_best  = v6_results[best_v6_name]['acc']
preds_v6     = v6_results[best_v6_name]['preds']

print(f'\n  Best v6 : {best_v6_name}  ->  {acc_v6_best:.4f}')
print(f'  vs v4 Dual-CBAM : {acc_v6_best - best_dual:+.4f}')
print(f'  vs v5 Frozen ViT: {acc_v6_best - acc_stack_trip:+.4f}')
print(f'  vs v3 baseline  : {acc_v6_best - 0.8641:+.4f}')
````

### 30 · v6 Classification Report & 3-Way Confusion Matrix

````python
print(f'=== v6 Triple Ensemble ({best_v6_name}) — Test Acc: {acc_v6_best:.4f} ===')
print(classification_report(y_test, preds_v6, target_names=CLASSES))

# 3-way side-by-side: v4 / v5 / v6
dual_pred_map2 = {'Soft': np.argmax(avg_probs,1),
                  'Weighted': np.argmax(wt_probs,1),
                  'Stack(dual)': preds_stack_dual}
pred_v4 = dual_pred_map2[best_dual_lbl]

fig, axes = plt.subplots(1, 3, figsize=(19, 5))
for ax, pred, title, cmap in zip(
        axes,
        [pred_v4, preds_trip, preds_v6],
        [f'v4 Dual-CBAM ({best_dual_lbl})\n{best_dual:.4f}',
         f'v5 Frozen ViT ({best_trip_name})\n{acc_stack_trip:.4f}',
         f'v6 FT-ViT ({best_v6_name})\n{acc_v6_best:.4f}'],
        ['Blues', 'Purples', 'Greens']):
    cm_n = confusion_matrix(y_test, pred).astype('float')
    cm_n /= (cm_n.sum(axis=1, keepdims=True) + 1e-8)
    sns.heatmap(cm_n, annot=True, fmt='.2f', cmap=cmap,
                xticklabels=CLASSES, yticklabels=CLASSES,
                linewidths=0.5, ax=ax)
    ax.set_title(title, fontweight='bold')
    ax.set_xlabel('Predicted'); ax.set_ylabel('True')
    plt.setp(ax.get_xticklabels(), rotation=45)

plt.suptitle('v4 vs v5 vs v6 Confusion Matrices', fontweight='bold', fontsize=13)
plt.tight_layout()
plt.savefig('/content/v6_triple_cm_comparison.png', dpi=120, bbox_inches='tight')
plt.show()
print('Saved: /content/v6_triple_cm_comparison.png')
````

### 31 · Per-Class Accuracy & Full Progression Table (v1 → v6)

````python
# ── Per-class accuracy 3-way ─────────────────────────────────────────────────
def per_class_acc(y_true, y_pred, classes):
    cm = confusion_matrix(y_true, y_pred)
    return {c: cm[i,i]/max(1,cm[i].sum()) for i,c in enumerate(classes)}

pca_v4 = per_class_acc(y_test, pred_v4,    CLASSES)
pca_v5 = per_class_acc(y_test, preds_trip, CLASSES)
pca_v6 = per_class_acc(y_test, preds_v6,   CLASSES)

df_pc = pd.DataFrame({
    'Class':     CLASSES,
    'v4_Dual':   [pca_v4[c] for c in CLASSES],
    'v5_FrzViT': [pca_v5[c] for c in CLASSES],
    'v6_FTViT':  [pca_v6[c] for c in CLASSES],
})
df_pc['Δ_v4_v6'] = df_pc['v6_FTViT'] - df_pc['v4_Dual']
df_pc['Δ_v5_v6'] = df_pc['v6_FTViT'] - df_pc['v5_FrzViT']
print('\n📋 Per-class accuracy: v4 / v5 / v6')
print(df_pc.to_string(index=False, float_format='{:.4f}'.format))

# Grouped bar chart
x = np.arange(len(CLASSES)); bw = 0.25
fig, ax = plt.subplots(figsize=(10, 4))
ax.bar(x-bw,  df_pc['v4_Dual'],   bw, label='v4 Dual-CBAM',           color='#1D6E8A', alpha=0.85)
ax.bar(x,     df_pc['v5_FrzViT'], bw, label='v5 Frozen ViT',          color='#8A1D6E', alpha=0.85)
ax.bar(x+bw,  df_pc['v6_FTViT'],  bw, label=f'v6 FT-ViT ({best_v6_name})', color='#1D8A44', alpha=0.85)
ax.set_xticks(x); ax.set_xticklabels(CLASSES, rotation=30)
ax.set_ylabel('Accuracy'); ax.set_ylim(0, 1.05)
ax.set_title('Per-class: v4 / v5 / v6', fontweight='bold')
ax.legend(); ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('/content/v6_perclass_comparison.png', dpi=120, bbox_inches='tight')
plt.show()

# ── Full progression table v1 → v6 ───────────────────────────────────────────
prog_rows = [
    ('v1', 'VGG16 (broken)',             0.3417),
    ('v2', 'VGG16',                      0.7759),
    ('v2', 'Soft ensemble',              0.7860),
    ('v3', 'VGG16',                      0.8548),
    ('v3', 'Stack ensemble',             0.8641),
    ('v4', 'VGG16_CBAM',                 RESULTS['VGG16_CBAM']['test_acc']),
    ('v4', 'ResNet50_CBAM',              RESULTS['ResNet50_CBAM']['test_acc']),
    ('v4', f'Dual-CBAM ({best_dual_lbl})',   best_dual),
    ('v5', f'Frozen ViT ({best_trip_name})', acc_stack_trip),
    ('v6', f'FT-ViT ({best_v6_name})',       acc_v6_best),
]
df_prog = pd.DataFrame(prog_rows, columns=['Version','Model','Acc'])
df_prog['Delta'] = df_prog['Acc'].diff().map(
    lambda x: f'{x:+.4f}' if pd.notna(x) else '—')
df_prog['Acc'] = df_prog['Acc'].map('{:.4f}'.format)
print('\n📊 FULL PROGRESSION — v1 to v6')
print(df_prog.to_string(index=False))

best_acc_val = float(df_prog['Acc'].astype(float).max())
best_r = df_prog[df_prog['Acc'].astype(float) == best_acc_val].iloc[0]
print(f'\n🏆 Best: {best_r["Version"]} — {best_r["Model"]} — {best_r["Acc"]}')

print('\n' + '='*65)
print('KEY FINDINGS')
print('='*65)
print(f'  v4 Dual-CBAM  : {best_dual:.4f}  (CNN-only baseline)')
print(f'  v5 Frozen ViT : {acc_stack_trip:.4f}  ({acc_stack_trip-best_dual:+.4f}) domain mismatch hurts')
print(f'  v6 FT ViT     : {acc_v6_best:.4f}  ({acc_v6_best-best_dual:+.4f}) after domain adaptation')
if acc_v6_best > best_dual:
    print('  ✅ Fine-tuning reversed the ViT degradation — domain adaptation worked!')
    print(f'  Net gain from ViT over CNN-only: {acc_v6_best-best_dual:+.4f}')
else:
    gap = acc_v6_best - best_dual
    print(f'  ViT fine-tuning improved over frozen ({acc_v6_best-acc_stack_trip:+.4f})')
    print(f'  Gap to dual-CBAM: {gap:+.4f} — try unfreezing 3 blocks or more epochs.')
print('='*65)
````
